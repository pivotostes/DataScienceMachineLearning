```python
import seaborn as sns
import numpy as np
```

```python
%matplotlib inline
```

```python
tips =  sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow', hue='sex')
```

```python
sns.boxplot(data=tips, palette='rainbow', orient='h')
```

```python
sns.violinplot(x='day', y='tip', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y=tips.total_bill, data=tips, jitter=True, hue=tips.sex, dodge=True)
```

```python
sns.swarmplot(x=tips.day, y=tips.total_bill, data=tips, hue=tips.sex, dodge=True)
```

```python
sns.swarmplot(x=tips.day, y=tips.total_bill, data=tips, color='black')
sns.violinplot(x=tips.day, y=tips.total_bill, data=tips)
```

```python
sns.catplot(x=tips.sex, y=tips.total_bill, data=tips, kind='bar')
```

```python
sns.catplot(x=tips.sex, y=tips.total_bill, data=tips, kind='bar')
```

```python
sns.catplot(x=tips.day, y=tips.total_bill, data=tips, kind='violin')
```
