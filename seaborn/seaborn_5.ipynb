```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head(20)
```

```python
iris.info()
```

```python
iris['species'].value_counts()
```

```python
grd = sns.PairGrid(iris)
```

```python
grd.map(plt.scatter)
```

```python
grd2 = sns.PairGrid(iris)
grd2.map_diag(plt.hist)
grd2.map_upper(plt.scatter)
grd2.map_lower(sns.kdeplot)
```
