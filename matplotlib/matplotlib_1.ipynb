```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x*x
```

```python
y
```

```python
# Funcional - forma mais simples de plotar um gráfico
plt.plot(x, y, '--r')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Gráfico X vs Y');
```

```python
# plt.subplots(nrows=1, ncols=2);
plt.subplot(1, 2, 1)
plt.plot(x, y, '--r')
plt.subplot(1, 2, 2)
plt.plot(y, x, '-og');
```

```python
fig = plt.figure()
axe1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axe2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

axe2.plot(y, x);
axe2.set_xlabel('Eixo_Y')
axe2.set_title('Gráfico Interno');

axe1.plot(x, y);
axe1.set_xlabel('Eixo_X')
axe1.set_title('Gráfico');
```
