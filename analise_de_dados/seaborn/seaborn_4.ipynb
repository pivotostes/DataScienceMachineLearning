```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex');
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v']);
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'], scatter_kws={'s':100});
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='sex', scatter_kws={'s':100});
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='day',aspect=0.6, height=8);
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='sex', row='time', scatter_kws={'s':100});
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='day',
           scatter_kws={'s':100}, aspect=0.6, height=10);
```
