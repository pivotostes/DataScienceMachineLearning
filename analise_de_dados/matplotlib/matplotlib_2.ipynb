```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x*x
```

```python
x
```

```python
y
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**3, 'b--')
ax.plot(x, x**4, 'g-.')
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_title('Gráfico x^3 e x^4');
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
```

```python
for axe in ax:
    axe.plot(x, y, 'b')
    axe.set_title('Titulo')
fig
```

```python
ax[0].plot(x, y, 'b')
ax[0].set_title('Titulo 1')
ax[1].plot(x, y, 'b')
ax[1].set_title('Titulo 2')
fig
```

```python
fig, ax = plt.subplots(nrows=5, ncols=5)
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8, 4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(8, 8))
axes.plot(x, y, 'r-', label='x²')
axes.plot(x, x**3, 'g--', label='x³')
axes.set_title('Título');
axes.legend(loc=0)
```

```python
axes
```

```python
fig.savefig('plot_1.png')
```
